## Parameter evaluation

### Basic Metrics

```python
strict_accuracy = (df["answer_ok"] == "yes").mean()
lenient_success_rate = df["answer_ok"].isin(["yes", "partially"]).mean()

overall_metrics = pd.DataFrame({
    "metric": ["Strict accuracy", "Lenient success rate"],
    "value": [round(strict_accuracy * 100, 1), round(lenient_success_rate * 100, 1)]
})
```

```python
display(overall_metrics)
```

### Per-Type Metrics

```python
per_type = df.groupby("expected_type")["answer_ok"].value_counts().unstack(fill_value=0)

for col in ["yes", "partially", "no"]:
    if col not in per_type.columns:
        per_type[col] = 0

per_type = per_type[["yes", "partially", "no"]]
per_type["total"] = per_type.sum(axis=1)
per_type["strict_accuracy (%)"] = (per_type["yes"] / per_type["total"] * 100).round(1)
per_type["lenient_success_rate (%)"] = ((per_type["yes"] + per_type["partially"]) / per_type["total"] * 100).round(1)
```

```python
display(per_type)
```

### Overall Label Distribution

```python
label_distribution = df["answer_ok"].value_counts().reindex(["yes", "partially", "no"], fill_value=0)
label_distribution_df = pd.DataFrame({
    "count": label_distribution,
    "percent (%)": (label_distribution / len(df) * 100).round(1)
})
```

## Evaluation Summary

- **Overall, the v2 workflow version performed reasonably well on the 20 evaluation cases.**  
  The system reached a **strict accuracy of 65.0%** and a **lenient success rate of 90.0%**, which means most questions could receive at least usable answers, while fully correct answers were achieved in about two-thirds of the test cases.

- **SQL queries were the strongest part of the system.**  
  The system handled structured database questions very reliably, including order counts, average payment value, customer distribution, and ranking-related questions. This shows that the SQL workflow is currently the most stable component.

- **Standard RAG questions were mostly usable, but answer quality was not always consistent.**  
  In most review-related questions, the system could identify complaint themes and customer sentiment correctly. However, some answers were too generic, repetitive, or lacked detail, so the RAG path still needs improvement in summarization quality.

- **Hybrid questions showed mixed performance.**  
  Some hybrid queries successfully combined price constraints with customer review information and returned useful suggestions. However, other hybrid cases struggled when the question required more precise category-level reasoning or stronger combination of SQL results and review evidence.

- **LLM fallback worked well for general and out-of-scope questions.**  
  For greetings, general knowledge, and unrelated requests, the system correctly avoided forcing SQL or RAG outputs and produced reasonable fallback answers.

- **Memory worked in short and explicit follow-up scenarios, but was less stable in longer conversations.**  
  When the follow-up question clearly referred to the previous answer, the system could use conversational context correctly. However, in long multi-turn conversations, memory performance dropped, especially when the follow-up wording was ambiguous.

- **Overall, the system performed best on SQL and fallback tasks.**  
  Standard RAG questions were generally usable, while hybrid and memory-related scenarios showed more variability and still need improvement in more complex cases.
